# Creamos los graficos correspondientes a cada minera

```python
plt.figure(figsize=(10,6))
x=[2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
y=[1103, 1086, 817, 1075, 1193, 1165, 1152, 1002, 925, 1242]
plt.plot(x, y, marker='o', linestyle='--', color='g', label='Escondida')
plt.xlabel('Años')
plt.ylabel('Empresa')
plt.title('Años vs Empresa')
plt.legend(loc='lower right')
plt.xticks([2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018])
plt.savefig('ejemplo.png')
```

```python
plt.figure(figsize=(10,6))
x=[2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
y=[574, 528, 443, 355, 339, 340, 308, 302, 330, 320]
plt.plot(x, y, marker='o', linestyle='--', color='g', label='Chuquicamata (1)')
plt.xlabel('Años')
plt.ylabel('Empresa')
plt.title('Años vs Empresa')
plt.legend(loc='lower right')
plt.xticks([2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018])
plt.savefig('ejemplo.png')
```
